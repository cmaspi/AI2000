```python
import random
import csv
import numpy as np
```

```python
def retList(filename):
    f=open(filename,'r')
    dataString = f.readlines()
    data=[]
    for i in range(len(dataString)):
        data.append(dataString[i].split())
    return np.array(data,dtype=float)

def accuracy(label, predicted):
    N=len(predicted)
    crt=0
    for i in range(N):
        if label[i] == predicted[i]:
            crt = crt+1
            if N != 0:
                return 0
            else :
                return crt/N
            
```

```python
def giniIndex(data):
    pos, neg = 0,0
    for i in data:
        if i[-1] == 0:
            neg = neg + 1
        else :
            pos = pos + 1
    pos = pos/(pos+neg)
    neg = 1- pos

    return 1 - pos*pos - neg*neg

def BestSplit(data, usedFeature, numThresholds):
    infoGain = 1e-9
    Left=[]
    Right=[]
    Threshold = None
    Feature = None
    n = len(data[0])-1

    for feature in range(n):
        if usedFeature[feature]==1:
            continue
        data = np.array(data)
        allValsinFeature = data[:,feature]
        minVal = np.min(allValsinFeature)
        maxVal = np.max(allValsinFeature)
        Tvals = np.linspace(minVal,maxVal,numThresholds+1,endpoint=False)[1:]

        for threshold in Tvals:
            left=[]
            right=[]
            for inputs in data:
                if inputs[feature] < threshold:
                    left.append(inputs)
                else:
                    right.append(inputs)
            if len(left)>0 and len(right)>0:
                delGini = giniIndex(data) - len(left)/len(data)*giniIndex(left)-len(right)/len(data)*giniIndex(right)
                if delGini > infoGain:
                    infoGain = delGini
                    Left=left
                    Right=right
                    Threshold = threshold
                    Feature = feature
    usedFeature[Feature]=1
    return {'threshold':Threshold, 'feature':Feature, 'left':Left, 'right':Right, 'usedFeature':usedFeature}

```

```python
class DecisionTree():
    tree={}

    def learn(self,data, usedFeature=None ,numThresholds=10):
        tree={'feature':None, 'value':None, 'threshold':None, 'usedFeature':[],'leftTree':None,'rightTree':None}
        if usedFeature is None:
            usedFeature= np.zeros(len(data[0])-1)
        if len(data)>0:
            if  giniIndex(data)<0.01:
                tree['usedFeature']=np.zeros(len(data[0])-1)
                best_split = BestSplit(data,tree['usedFeature'],numThresholds)
                tree['feature']=best_split['feature']
                tree['threshold']=best_split['threshold']
                leftTree={}
                rightTree={}
                leftTree=DecisionTree.learn(self,best_split['left'],best_split['usedFeature'],numThresholds)
                rightTree=DecisionTree.learn(self,best_split['right'],best_split['usedFeature'],numThresholds)
                tree['leftTree']=leftTree   
                tree['rightTree']=rightTree
                self.tree = tree
                return tree
            else:
                tree['leftTree']=None
                tree['rightTree']=None
                labels=list(data[-1])
                tree['value']=max(labels,key=labels.count)
                return tree
    def classify(self, tree, instance):
        if tree['value'] is not None:
            return tree['value']
        else:
            f = instance[tree['feature']]
            if f<tree['threshold']:
                return self.classify(tree['leftTree'],instance)
            else:
                return self.classify(tree['rightTree'],instance)

def runDecisionTree(train, test):
    Tree= DecisionTree()
    tree = Tree.learn(train)
    results=[]
    for instance in test:
        result = Tree.classify(tree,instance[:-1])
        print(result)
        results.append(result == instance[-1])
    accuracy = float(results.count(True))/float(len(results))
    print(accuracy)
```
